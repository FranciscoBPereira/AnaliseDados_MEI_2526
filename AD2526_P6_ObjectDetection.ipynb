**1. Model Downloading**

```python
from ultralytics import YOLO

# Load a light YOLO version released in 2023
model = YOLO("yolov8n.pt")
```

```python
# Model Information

model.info()
```

```python
# Model architecture

print(model.model)
```

**2. Object Detection for a Single Image**

```python
# Upload a new image to the working directory
# Check how the model behaves

import matplotlib.pyplot as plt
from PIL import Image

# Load image
Image_Name = 'A.jpeg' #@param {type:"string"}


# Appy YOLO model
results = model(Image_Name)

original = Image.open(Image_Name)


annotated = results[0].plot()
annotated_rgb = annotated[..., ::-1]

# Print Images
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].imshow(original)
axes[0].set_title("Original")
axes[0].axis("off")

axes[1].imshow(annotated_rgb)
axes[1].set_title("Detections")
axes[1].axis("off")

plt.show()
```

```python
for r in results:
    for box in r.boxes:
        class_id = int(box.cls[0])
        class_name = model.names[class_id]

        confidence = float(box.conf[0])
        bbox = box.xyxy[0]

        print(class_name)
        print(confidence)
        print(bbox)
```

**3. Object Detection for Multiple Images**

```python
# Create two directories in the working folder
# and specify the names on the left

# Upload several images to the input directory

# YOLO is applied to all images
# The result is saved to the output directory

import os

input_dir = 'input' #@param {type:"string"}

output_dir = 'output' #@param {type:"string"}

for filename in os.listdir(input_dir):
    if filename.lower().endswith((".jpg", ".jpeg", ".png")):
        image_path = os.path.join(input_dir, filename)

        # Ler imagem
        original = Image.open(image_path)

        # Deteção
        results = model(image_path)

        # Obter imagem anotada
        annotated_bgr = results[0].plot()
        annotated_rgb = annotated_bgr[..., ::-1]  # BGR -> RGB

        # Criar figura lado a lado
        fig, axes = plt.subplots(1, 2, figsize=(14, 7))

        # Original
        axes[0].imshow(original)
        axes[0].set_title("Original")
        axes[0].axis("off")

        # Anotada
        axes[1].imshow(annotated_rgb)
        axes[1].set_title("Detections with Labels")
        axes[1].axis("off")


        # Mostrar figura
        plt.show()

        # Guardar figura lado a lado
        save_path = os.path.join(output_dir, f"result_{filename}")
        fig.savefig(save_path)
        plt.close(fig)
```
